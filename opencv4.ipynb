```python
import cv2
import numpy as np
```

## colour object detection

```python
cap = cv2.VideoCapture(0)
def funct(x):
    pass
cv2.namedWindow("demo")
cv2.createTrackbar("lb","demo",0,255,funct)
cv2.createTrackbar("lg","demo",0,255,funct)
cv2.createTrackbar("lr","demo",0,255,funct)

cv2.createTrackbar("ub","demo",255,255,funct)
cv2.createTrackbar("ug","demo",255,255,funct)
cv2.createTrackbar("ur","demo",255,255,funct)

while True:
    r,frame = cap.read()
    if r == True:
        img = cv2.resize(frame, (400,300))
        hsv_img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
        Lb = cv2.getTrackbarPos("lb","demo")
        Lg = cv2.getTrackbarPos("lb","demo")
        Lr = cv2.getTrackbarPos("lb","demo")
        Ub = cv2.getTrackbarPos("lb","demo")
        Ug = cv2.getTrackbarPos("lb","demo")
        Ur = cv2.getTrackbarPos("lb","demo")
        lo = np.array([Lb,Lg,Lr])
        up = np.array([Ub,Ug,Ur])
        masks = cv2.inRange(hsv_img, lo, up)
        res = cv2.bitwise_and(img, img, mask=masks)
        cv2.imshow("res",res)
        cv2.imshow("mask",masks)
        cv2.imshow("hsv",hsv_img)
        cv2.imshow("window",img)
        if cv2.waitKey(1) & 0xff == ord("p"):
            break
            
    else:
        break
    
cap.release()
cv2.destroyAllWindows()
cv2.waitKey(1) 

        
        
```

## threshold

```python
img = cv2.imread("img1.jpg")
img = cv2.resize(img, (500,600))
_,thr = cv2.threshold(img,100,255,cv2.THRESH_BINARY)
cv2.imshow("threshold",thr)
cv2.imshow("window",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.waitKey(1)
```

## histogram of the b, g, r images

```python
img = cv2.imread("img1.jpg")
img = cv2.resize(img, (500,600))

b,g,r = cv2.split(img)
plt.figure(figsize=(10,10))

hist1 = cv2.calcHist([b],[0],None,[256],[0,256])
hist2 = cv2.calcHist([g],[0],None,[256],[0,256])
hist3 = cv2.calcHist([r],[0],None,[256],[0,256])

plt.subplot(3,1,1)
plt.plot(hist1)

plt.subplot(3,1,2)
plt.plot(hist2)

plt.subplot(3,1,3)
plt.plot(hist3)

plt.show()

cv2.imshow("img", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.waitKey(1)
```

```python
img = cv2.imread("img1.jpg")
img = cv2.resize(img, (500,600))

b,g,r = cv2.split(img)
plt.figure(figsize=(8,8))

plt.hist(b.ravel(), 256, [0, 256], color='blue')
plt.hist(g.ravel(), 256, [0, 256], color='green')
plt.hist(r.ravel(), 256, [0, 256], color='red')
##b is the Blue channel of the image (2D NumPy array) .ravel() flattens the 2D array into a 1D array.


plt.show()

cv2.imshow("img", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.waitKey(1)
```
