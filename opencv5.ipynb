## contour of an image


```python
img = cv2.imread("img1.jpg")
img = cv2.resize(img, (500,600))
img1 = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

r,th = cv2.threshold(img1,100,255,cv2.THRESH_BINARY)
c,h = cv2.findContours(th,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
 
new_img = cv2.drawContours(img,c,-1,(255,0,0),2)

cv2.imshow("window",new_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.waitKey(1)
```

## convexHull contour

```python
img = cv2.imread("img1.jpg")
img = cv2.resize(img, (500,600))
img1 = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

r,th = cv2.threshold(img1,100,255,cv2.THRESH_BINARY)

cnt,h = cv2.findContours(th,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
arr = []
for c in cnt:
    m = cv2.moments(c)
    if m["m00"] == 0:
        continue                   # skip this contour to avoid division by zero

    x = int(m["m10"] / m["m00"])     ## Calculates the centroid:
    y = int(m["m01"] / m["m00"])
    cv2.drawContours(img, cnt, -1, (0,0,255), 4)
    
    cv2.circle(img,(x,y),5,(255,0,0),-1) ## point to show the centre of the object
    
    a = cv2.contourArea(c)
    arr.append(a)                  ## Computes the area of the current contour and stores it in a list arr.
    
    ep = 0.1*cv2.arcLength(c,True)
    d = cv2.approxPolyDP(c,ep,True)   ## Approximate Contour Shape
    
    h = cv2.convexHull(d)     # cv2.convexHull() finds the outermost boundary of this shape — removing any indentations or concavities.
    
    x,y,w,h = cv2.boundingRect(h)
    cv2.rectangle(img, (x,y), (x+w, y+h), (0,255,0), 5)
    

print(arr)
print(cnt)
cv2.imshow("window",th)
cv2.imshow("window2",img1)
cv2.imshow("window1",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.waitKey(1)
```
